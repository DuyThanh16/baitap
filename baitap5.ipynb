```python
#w5a3
import math
def sont(n:int) -> bool:
    for i in range(2, math.isqrt(n)+1):
        if n%i==0:
            return False
    return (n>=2)
n=int(input())
for x in range(2, n+1):
    if(sont(x)):
        print(x, end=" ")
```

```python
#w5a5
def vt(arr, k):
    for i in range(len(arr)):
        if arr[i]==k:
            return i+1
    return -1
arr=list(map(int,input().split()))
k=int(input())
print(vt(arr,k))
```

```python
#w5a6
n=int(input())
def gt(n):
    gt=1
    for i in range(1, n+1):
        gt*=i
    return gt
print(gt(n))
```

```python
#w5a9
def tong_chu_so(n):
    tong=0
    while n>0:
        tong+=n%10
        n//=10
    return tong
n=int(input())
print(tong_chu_so(n))
```
