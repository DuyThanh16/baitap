```python
#W7A1
def binary_search(arr, left, right, target):
    if left > right:
        return -1

    mid = (left + right) // 2

    if arr[mid] == target:
        return mid
    elif target < arr[mid]:
        return binary_search(arr, left, mid - 1, target)
    else:
        return binary_search(arr, mid + 1, right, target)

arr=list(map(int,input().split()))
target=int(input())
result = binary_search(arr, 0, len(arr) - 1, target)
print(result)
```

```python
#W7A2
def count_occurrences(arr, x) -> int:
    cnt=0
    for i in arr:
        if i == x:
            cnt+=1
    return cnt
arr=list(map(int,input().split()))
x=int(input())
print(count_occurrences(arr, x))
```

```python
# Viết hàm tìm phần tử có số
# lần xuất hiện nhiều nhất
# trong danh sách. Nếu có
# nhiều phần tử cùng số lần,
# trả về phần tử xuất hiện
# trước.
def most_frequent_first(lst):
    freq = {}
    max_count = 0
    result = None
    for x in lst:
        freq[x] = freq.get(x, 0) + 1
        if freq[x] > max_count:
            max_count = freq[x]
            result = x
    return result
lst = list(map(int, input().split()))
print( most_frequent_first(lst))
```

```python
# Bạn được cung cấp một
# danh sách các danh sách
# lồng nhau, vd [[1, 2, 3], [4,
# 5], [6, 7, 8, 9]], hãy làm
# phẳng danh sách
nested_list = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

flat_list = []
for sublist in nested_list:
    for item in sublist:
        flat_list.append(item)

print(flat_list)
```

```python
# Cho một mảng số nguyên không âm. Gọi
# mảng đó là "mảng đẹp" nếu mỗi phần tử
# x trong mảng xuất hiện đúng x lần.
# Kiểm tra xem mảng có phải là mảng đẹp
# hay không. Nếu không hãy xóa đi y số (y
# là nhỏ nhất có thể) để mảng còn lại thành
# mảng đẹp.
from collections import Counter

n = int(input())
arr = list(map(int, input().split()))

freq = Counter(arr)
deletions = 0
is_beautiful = True

for x, count in freq.items():
    if count != x:
        is_beautiful = False
    if count > x:
        deletions += count - x
    elif count < x:
        deletions += count

if is_beautiful:
    print("YES")
else:
    print("NO")

print(deletions)
```
