```python
#w4a3
#Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n)
# với 0 < n < 100. 
n = int(input("Nhap so nguyen  n (0 < n < 100): "))
if 0 < n < 100:
    s = 1
    for i in range(1, n + 1):
        s *= i
    print(f"{n}! = {s}")
else:
    print("Nhap so 0 < n < 100")
```

```python
#w4a4
#Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu)
#của số đó. Không dùng kiểu dữ liệu string.
n = int(input("Nhap so nguyen n: "))
n = abs(n)
if n == 0:
    dem = 1
else:
    dem = 0
    while n > 0:
        n //= 10     
        dem += 1    

print("So chu so la:", dem)
```

```python
# Hãy viết một chương trình nhận vào một số nguyên dương n và một dãy gồm n số
# nguyên. Nếu n số nguyên đó có số 42 thì in ra màn hình dòng chữ "I′ve found the
# meaning of life!", ngược lại in ra dòng chữ "It′s a joke!".
n = int(input("Nhap so nguyen duong n: "))
numbers = []
for i in range(n):
    x = int(input(f"Nhap so thu {i+1}: "))
    numbers.append(x)
if 42 in numbers:
    print("I've found the meaning of life!")
else:
    print("It's a joke!")
```

```python
# Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
# [a, b]. Mỗi dòng là một bộ test (a b)."
def is_prime(n):
    """Kiểm tra n có phải là số nguyên tố không"""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


while True:
    try:
        line = input("Nhập a b (hoặc nhấn Enter để kết thúc): ").strip()
        if line == "":
            break  # dừng khi người dùng không nhập gì
        a, b = map(int, line.split())
        if a > b or a <= 0:
            print("Vui lòng nhập a ≤ b và cả hai là số nguyên dương!")
            continue

        tong = 0
        for i in range(a, b + 1):
            if is_prime(i):
                tong += i

        print("Tổng các số nguyên tố trong đoạn [{}, {}] là: {}".format(a, b, tong))

    except ValueError:
        print("Dữ liệu không hợp lệ! Hãy nhập hai số nguyên cách nhau bởi dấu cách.")
```

```python
# Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
# tìm ước chung lớn nhất của hai số m và n. 
m = int(input("Nhập số nguyên dương m: "))
n = int(input("Nhập số nguyên dương n: "))
if m <= 0 or n <= 0:
    print("Vui lòng nhập hai số nguyên dương")
else:
    # Thuật toán Euclid
    while n != 0:
        m, n = n, m % n
    print("Ước chung lớn nhất là:", m)
```
