```python
n=int(input())
rev=0
while n>0:
    rev=rev*10+n%10
    n//=10
print(rev)
```

```python
def tong_chu_so(n:int)->int:
    tong=0
    while n>0:
        tong+=n%10
        n//=10
    return tong
n=int(input())
print(tong_chu_so(n))
```
