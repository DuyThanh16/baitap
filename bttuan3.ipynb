```python
# Kiểm tra xem một số có phải là lũy thừa của 2 hay không
# Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2
# hay không. 
n = int(input())

if n > 0 and (n & (n - 1)) == 0:
    print("La luy thua cua 2")
else:
    print("Khong phai luy thua cua 2")
```

```python
# Viêt chương trình chia một số m cho n, kết quả thu được làm tròn lên
m = int(input("Nhap m: "))
n = int(input("Nhap n: "))

ket_qua = (m + n - 1) // n
print(ket_qua)
```

```python
# Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm
# tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in
# ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều",
# "Tam giác cân", hay "Tam giác thường".
a = int(input())
b = int(input())
c = int(input())
if a + b <= c or a + c <= b or b + c <= a:
    print("Không phải tam giác")
else:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or a == c or b == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
```

```python
# Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra
# nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô
# nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.
# Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra
# nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô
# nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.
a = float(input())
b = float(input())
if a == 0:
    if b == 0:
        print("Vô số nghiệm")
    else:
        print("Vô nghiệm")
else:
    x = -b / a
    print(f"{x:.2f}")
```

```python
# Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số
# nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả
# cách nhau một dấu cách
x = float(input())
nguyen = int(x)
if x >= 0:
    tron_xuong = nguyen
else:
    tron_xuong = nguyen if x == nguyen else nguyen - 1
if x >= 0:
    tron_len = nguyen if x == nguyen else nguyen + 1
else:
    tron_len = nguyen
phan_du = abs(x - nguyen)
if phan_du < 0.5:
    gan_nhat = nguyen
else:
    if x > 0:
        gan_nhat = nguyen + 1
    else:
        gan_nhat = nguyen - 1
print(tron_len, tron_xuong, gan_nhat)
```
