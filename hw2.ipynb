```python
#7-hw2
x = int(input())    
if x < 0:
    print("Vui lòng nhập một số không âm.")
else:
    gio = x // 3600
    phut = (x % 3600) // 60
    giay = x % 60

    print(f"{x} giây = {gio} giờ, {phut} phút, {giay} giây")

```
