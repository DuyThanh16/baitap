```python
# Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ
# thuật toán sắp xếp nào 
def merge_sorted_lists(a, b):
    i = j = 0
    result = []

    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            result.append(a[i])
            i += 1
        else:
            result.append(b[j])
            j += 1


    result.extend(a[i:])
    result.extend(b[j:])

    return result
a = [1, 2, 3, 4, 5 ]
b = [2, 3, 6, 7, 8, 9 ]


print(merge_sorted_lists(a, b))
```

```python
# Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ
# số để được một số nguyên nhỏ nhất có thể 
n = input().strip()
digits = list(n)
length = len(digits)

for i in range(length):
    min_digit = digits[i]
    min_pos = -1

    # tìm chữ số nhỏ nhất bên phải (ưu tiên vị trí ngoài cùng bên phải)
    for j in range(length - 1, i, -1):
        if digits[j] < min_digit:
            min_digit = digits[j]
            min_pos = j

    if min_pos != -1:
        digits[i], digits[min_pos] = digits[min_pos], digits[i]
        break

print(int("".join(digits)))

```

```python
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        n = f.read().strip()
    return int(n[::-1])
```
